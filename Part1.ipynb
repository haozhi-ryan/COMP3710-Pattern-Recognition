```python
# Task 1.1
import torch
import numpy as np

print("PyTorch Version:", torch.__version__)

# --- Device configuration ---
# Check if CUDA is available, otherwise set device to CPU
# CUDA is NVIDIA's parallel computing platform that lets software use a GPU for computing
# This will allow the code to run on a GPU if available, which is much fast thern a CPU for image proccessing,
# simulations, and big matrix math since it can run thousands of small operations in parallel.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# --- Grid ---
# Grid for computing image, subdivide the space
# Creates a grid from -4.0 to 4.0, with a step size of 0.01
# X and Y are 2D arrays representing the grid coordinates
X, Y = np.mgrid[-4.0:4:0.01, -4.0:4:0.01]

# --- load into PyTorch tensors ---
# Converts the NumPy arrays X and Y into PyTorch tensors 
# A tensor is a multi-dimensional array data structure
# Allows running operations on the GPU or CPU, making it better for image processing,
# simulations, and big matrix math since it can run thousands of small operations in parallel.
x = torch.Tensor(X)
y = torch.Tensor(Y)

# --- Transfer to the device ---
# Moves the tensors x and y to the location specified by the device
# If device = "cpu" → keeps them in system memory (normal CPU processing).
# If device = "cuda" → moves them into the GPU’s memory so PyTorch can run operations on the GPU.
x = x.to(device)
y = y.to(device)

# --- Compute Gaussian ---
# For each point in x and y, compute the Gaussian function
# The Gaussian function is a bell-shaped curve, often used in statistics and probability. 
z = torch.exp(-(x**2+y**2)/2.0)

# --- Plot --- 
import matplotlib.pyplot as plt

# Convert the output tensor z to a NumPy array.
# Displays that array as an image, with values mapped to colors.
# z.cpu().numpy() → moves the tensor z to the CPU (if it was on GPU) and converts it into a NumPy array so Matplotlib can plot it.
plt.imshow(z.cpu().numpy())

# Automatically adjusts spacing so nothing overlaps or gets cut off.
plt.tight_layout()
plt.show()
```

```python
# Task 1.2
# AI Prompt: Generate a Python script to plot a 2D Gaussian function using Numpy and Matplotlib

import numpy as np
import matplotlib.pyplot as plt

# --- Define the grid ---
x = np.linspace(-4, 4, 200) # Makes 200 evenly spaced numbers from -4 to 4.
y = np.linspace(-4, 4, 200) 

# Takes those 1D arrays x and y and builds two 2D arrays:
X, Y = np.meshgrid(x, y) # X contains the x-coordinates, Y contains the y-coordinates for each point in the grid.
# Each (X[i,j], Y[i,j]) pair is a coordinate on the grid.
# E.g. Point at row 2 and column 3 corresponds to (X[2,3], Y[2,3]) (0 is row 0)

# --- Define the 2D Gaussian function ---
# Applies the 2D Gaussian function to every coordinate in your grid.
Z = np.exp(-(X**2 + Y**2) / 2.0)

# --- Plot the Gaussian ---

# shows the Gaussian values in Z mapped to colors, with axes scaled from -4 to 4 in both X and Y.
# cmap='inferno' tells Matplotlib to use the Inferno colormap — a smooth gradient going from dark purple → blue → green → yellow.
plt.imshow(Z, extent=[-4, 4, -4, 4], cmap='inferno')

# Adds a side bar showing how color corresponds to numeric values.
plt.colorbar(label='Intensity')
plt.title('2D Gaussian Function (NumPy)')

# Adds labels for x and y axis
plt.xlabel('X')
plt.ylabel('Y')

# Fixes spacing so labels fit nicely.
plt.tight_layout()

plt.show()
```
